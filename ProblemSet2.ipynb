b(): b() is called a third time. The current value of "a" is 4, (updated from the previous call). The function "c(a)" is called adding 2 to the current value of "a", returning 6. This value, 6 is now assigned back to the global variable "a".

Write function collatz() that takes a positive integer x as input and prints the Collatz sequence starting at x.

A Collatz sequence is obtained by repeatedly applying this rule to the previous number x in the sequence:

x = { 𝑥/2 𝑖𝑓 𝑥 𝑖𝑠 𝑒𝑣𝑒𝑛
3𝑥+1 𝑖𝑓 𝑥 𝑖𝑠 𝑜𝑑𝑑

Your function should stop when the sequence gets to number 1. Your implementation must be recursive, without any loops.

# Solution 4

```python
def collatz(x):
    print(x)
    if x == 1:
        return
    if x%2 == 0:
        collatz(x // 2)
    else: 
        collatz(3*x + 1)
print(collatz(10))
```

# Question 5

Write a recursive method binary() that takes a non-negative integer n and prints the binary representation of integer n. 
>>> binary(0) 0 
>>> binary(1) 1
>>> binary(3) 11
>>> binary(9) 1001

# solution 5

```python
def binary(n):
    if n == 0:
        return "0"
    elif n ==1:
        return "1"
    else: 
        return binary(n // 2) + str(n % 2)
print(binary(0))
print(binary(1))
print(binary(3))
print(binary(9))
```

# Question 7

Implement recursive function webdir() that takes as input: a URL (as a string) and non-negative integers depth and indent. Your function should visit every web page reachable from the starting URL web page in depth clicks or less, and print each web page's URL. As shown below, indentation, specified by indent, should be used to indicate the depth of a URL. >>> 

webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html', 2, 0)
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html 
http://reed.cs.depaul.edu/lperkovic/csc242/test2.html 
http://reed.cs.depaul.edu/lperkovic/csc242/test4.html 
http://reed.cs.depaul.edu/lperkovic/csc242/test3.html 
http://reed.cs.depaul.edu/lperkovic/csc242/test4.html

# Solution 7

```python
import requests
from bs4 import BeautifulSoup   #bs4 is the import statement to import the beautiful soup library(used for parsing html and xml docs)

def webdir(url, depth, indent):
    print(" " * indent + url)    #prints url with correct indentation
    
    #base case: reached maximum depth
    if depth == 0:
        return
    try: 
        #retrieve the web page content
        response = requests.get(url)
        content = response.text
        
        #create a BeautifulSoup object for HTML parsing
        soup = BeautifulSoup(content, "html.parser")
        
        #find all <a> tags for links
        links = soup.find_all("a")
        
        #visit each link recursively
        for link in links:
            href = link.get("href")
            
            #check if the href is a complete url or relative path
            if href.startswith("http://") or href.startswith("https://"):
                new_url = href
            else:
                
                #convert path to a url
                new_url = url + "/" + href
            
            #call recursively webdir with reduced depth and increased indentation
            webdir(new_url, depth - 1, indent + 4)
            
    except requests.exceptions.RequestException:
            print(" " * indent + "Error retrieving web page: ", url)

webdir("http://reed.cs.depaul.edu/lperkovic/csc242/test1.html", 2, 0)
print("http://reed.cs.depaul.edu/lperkovic/csc242/test2.html") 
print("http://reed.cs.depaul.edu/lperkovic/csc242/test4.html") 
print("http://reed.cs.depaul.edu/lperkovic/csc242/test3.html") 
print("http://reed.cs.depaul.edu/lperkovic/csc242/test4.html")

        
```

# Solution 9

```python
def words(*args):
    
    upperCaseWords = [word.upper() for word in args]
    lowerCaseWords = [word.lower() for word in args]
    wordLengths = [len(word) for word in args]
    wordDetails = [[word.upper(),word.lower(), len(word)] for word in args]
    longWords = [word for word in args if len(word)>=4]
    
    return  upperCaseWords, lowerCaseWords, wordLengths, wordDetails, longWords
result = words('The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog')
upperCaseWords, lowerCaseWords, wordLengths, wordDetails, longWords = result


#print(animal = "The","quick", "brown","fox","jumped","over","the","lazy","dog")
print(upperCaseWords)
print(lowerCaseWords)
print(wordLengths)
print(wordDetails)
print(longWords)
```
